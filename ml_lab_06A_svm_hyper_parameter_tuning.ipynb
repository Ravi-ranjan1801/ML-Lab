```python
# Install cvxopt (
!pip install cvxopt

import numpy as np
from cvxopt import matrix, solvers

# Step 1: Define dataset

X = np.array([
    [1,1],
    [2,2],
    [3,3],
    [4,4]
], dtype=float)

# Labels (+1, +1, -1, -1)
y = np.array([1, 1, -1, -1], dtype=float)

n_samples, n_features = X.shape

# Step 2: Construct QP matrices
# Variables = [w1, w2, b]

# P (quadratic term)
P = np.zeros((3,3))
P[0,0] = 1
P[1,1] = 1
P = matrix(P)

# q (linear term)
q = matrix(np.zeros(3))

# Step 3: Constraints Gx ≤ h
# Convert yi(w.x + b) ≥ 1 → -yi(w.x + b) ≤ -1
G = []
h = []

for i in range(n_samples):
    xi = X[i]
    yi = y[i]

    # [-yi*x1, -yi*x2, -yi]
    row = [-yi*xi[0], -yi*xi[1], -yi]
    G.append(row)
    h.append(-1)

G = matrix(np.array(G, dtype=float))
h = matrix(np.array(h, dtype=float))

# Step 4: Solve QP

sol = solvers.qp(P, q, G, h)

solution = np.array(sol['x']).flatten()

w1, w2, b = solution

# Step 5: Results

print("Optimal w1:", w1)
print("Optimal w2:", w2)
print("Optimal b :", b)

# Objective value = 1/2 (w1^2 + w2^2)
objective_value = 0.5 * (w1**2 + w2**2)

print("\nOptimal Objective Value:", objective_value)

# Step 6: Decision Boundary

print("\nDecision Boundary: {}*x + {}*y + {} = 0".format(w1, w2, b))
```
